```python

import mysql.connector
import csv
import re
# Conexão com o banco de dados MariaDB
db = mysql.connector.connect(
    host="localhost",
    user="root",
    password="richard2004",
    database="dbtrab",
    charset="utf8mb4",
    collation="utf8mb4_general_ci"
)

cursor = db.cursor()

# Mapeamento completo de países por continente
PAISES_CONTINENTES = {
    # Europa
    'Albania': 'Europa',
    'Andorra': 'Europa',
    'Armenia': 'Europa',
    'Austria': 'Europa',
    'Azerbaijan': 'Europa',
    'Belarus': 'Europa',
    'Belgium': 'Europa',
    'Bosnia and Herzegovina': 'Europa',
    'Bulgaria': 'Europa',
    'Croatia': 'Europa',
    'Cyprus': 'Europa',
    'Czech Republic': 'Europa',
    'Denmark': 'Europa',
    'England': 'Europa',
    'Estonia': 'Europa',
    'Faroe Islands': 'Europa',
    'Finland': 'Europa',
    'France': 'Europa',
    'Georgia': 'Europa',
    'Germany': 'Europa',
    'Gibraltar': 'Europa',
    'Greece': 'Europa',
    'Hungary': 'Europa',
    'Iceland': 'Europa',
    'Ireland': 'Europa',
    'Italy': 'Europa',
    'Kazakhstan': 'Europa',
    'Kosovo': 'Europa',
    'Latvia': 'Europa',
    'Liechtenstein': 'Europa',
    'Lithuania': 'Europa',
    'Luxembourg': 'Europa',
    'Malta': 'Europa',
    'Moldova': 'Europa',
    'Monaco': 'Europa',
    'Montenegro': 'Europa',
    'Netherlands': 'Europa',
    'North Macedonia': 'Europa',
    'Northern Ireland': 'Europa',
    'Norway': 'Europa',
    'Poland': 'Europa',
    'Portugal': 'Europa',
    'Romania': 'Europa',
    'Russia': 'Europa',
    'San Marino': 'Europa',
    'Scotland': 'Europa',
    'Serbia': 'Europa',
    'Slovakia': 'Europa',
    'Slovenia': 'Europa',
    'Spain': 'Europa',
    'Sweden': 'Europa',
    'Switzerland': 'Europa',
    'Turkey': 'Europa',
    'Ukraine': 'Europa',
    'Vatican City': 'Europa',
    'Wales': 'Europa',
    'Yugoslavia': 'Europa',
    'Soviet Union': 'Europa',
    'Czechoslovakia': 'Europa',
    'Basque Country': 'Europa',
    'Catalonia': 'Europa',

    # América do Sul
    'Argentina': 'América do Sul',
    'Bolivia': 'América do Sul',
    'Brazil': 'América do Sul',
    'Chile': 'América do Sul',
    'Colombia': 'América do Sul',
    'Ecuador': 'América do Sul',
    'Guyana': 'América do Sul',
    'Paraguay': 'América do Sul',
    'Peru': 'América do Sul',
    'Suriname': 'América do Sul',
    'Uruguay': 'América do Sul',
    'Venezuela': 'América do Sul',

    # América do Norte e Central
    'Anguilla': 'América do Norte',
    'Antigua and Barbuda': 'América do Norte',
    'Aruba': 'América do Norte',
    'Bahamas': 'América do Norte',
    'Barbados': 'América do Norte',
    'Belize': 'América do Norte',
    'Bermuda': 'América do Norte',
    'Canada': 'América do Norte',
    'Costa Rica': 'América do Norte',
    'Cuba': 'América do Norte',
    'Curaçao': 'América do Norte',
    'Dominica': 'América do Norte',
    'Dominican Republic': 'América do Norte',
    'El Salvador': 'América do Norte',
    'Grenada': 'América do Norte',
    'Guatemala': 'América do Norte',
    'Haiti': 'América do Norte',
    'Honduras': 'América do Norte',
    'Jamaica': 'América do Norte',
    'Mexico': 'América do Norte',
    'Nicaragua': 'América do Norte',
    'Panama': 'América do Norte',
    'Puerto Rico': 'América do Norte',
    'Saint Kitts and Nevis': 'América do Norte',
    'Saint Lucia': 'América do Norte',
    'Saint Vincent and the Grenadines': 'América do Norte',
    'Trinidad and Tobago': 'América do Norte',
    'United States': 'América do Norte',

    # Ásia
    'Afghanistan': 'Ásia',
    'Bahrain': 'Ásia',
    'Bangladesh': 'Ásia',
    'Bhutan': 'Ásia',
    'Brunei': 'Ásia',
    'Cambodia': 'Ásia',
    'China': 'Ásia',
    'East Timor': 'Ásia',
    'India': 'Ásia',
    'Indonesia': 'Ásia',
    'Iran': 'Ásia',
    'Iraq': 'Ásia',
    'Israel': 'Ásia',
    'Japan': 'Ásia',
    'Jordan': 'Ásia',
    'Kuwait': 'Ásia',
    'Kyrgyzstan': 'Ásia',
    'Laos': 'Ásia',
    'Lebanon': 'Ásia',
    'Malaysia': 'Ásia',
    'Maldives': 'Ásia',
    'Mongolia': 'Ásia',
    'Myanmar': 'Ásia',
    'Nepal': 'Ásia',
    'North Korea': 'Ásia',
    'Oman': 'Ásia',
    'Pakistan': 'Ásia',
    'Palestine': 'Ásia',
    'Philippines': 'Ásia',
    'Qatar': 'Ásia',
    'Saudi Arabia': 'Ásia',
    'Singapore': 'Ásia',
    'South Korea': 'Ásia',
    'Sri Lanka': 'Ásia',
    'Syria': 'Ásia',
    'Taiwan': 'Ásia',
    'Tajikistan': 'Ásia',
    'Thailand': 'Ásia',
    'Turkmenistan': 'Ásia',
    'United Arab Emirates': 'Ásia',
    'Uzbekistan': 'Ásia',
    'Vietnam': 'Ásia',
    'Yemen': 'Ásia',

    # África
    'Algeria': 'África',
    'Angola': 'África',
    'Benin': 'África',
    'Botswana': 'África',
    'Burkina Faso': 'África',
    'Burundi': 'África',
    'Cameroon': 'África',
    'Cape Verde': 'África',
    'Central African Republic': 'África',
    'Chad': 'África',
    'Comoros': 'África',
    'Congo': 'África',
    'DR Congo': 'África',
    'Djibouti': 'África',
    'Egypt': 'África',
    'Equatorial Guinea': 'África',
    'Eritrea': 'África',
    'Eswatini': 'África',
    'Ethiopia': 'África',
    'Gabon': 'África',
    'Gambia': 'África',
    'Ghana': 'África',
    'Guinea': 'África',
    'Guinea-Bissau': 'África',
    'Ivory Coast': 'África',
    'Kenya': 'África',
    'Lesotho': 'África',
    'Liberia': 'África',
    'Libya': 'África',
    'Madagascar': 'África',
    'Malawi': 'África',
    'Mali': 'África',
    'Mauritania': 'África',
    'Mauritius': 'África',
    'Morocco': 'África',
    'Mozambique': 'África',
    'Namibia': 'África',
    'Niger': 'África',
    'Nigeria': 'África',
    'Rwanda': 'África',
    'São Tomé and Príncipe': 'África',
    'Senegal': 'África',
    'Seychelles': 'África',
    'Sierra Leone': 'África',
    'Somalia': 'África',
    'South Africa': 'África',
    'South Sudan': 'África',
    'Sudan': 'África',
    'Tanzania': 'África',
    'Togo': 'África',
    'Tunisia': 'África',
    'Uganda': 'África',
    'Zambia': 'África',
    'Zimbabwe': 'África',

    # Oceania
    'Australia': 'Oceania',
    'Fiji': 'Oceania',
    'Kiribati': 'Oceania',
    'Marshall Islands': 'Oceania',
    'Micronesia': 'Oceania',
    'Nauru': 'Oceania',
    'New Zealand': 'Oceania',
    'Palau': 'Oceania',
    'Papua New Guinea': 'Oceania',
    'Samoa': 'Oceania',
    'Solomon Islands': 'Oceania',
    'Tonga': 'Oceania',
    'Tuvalu': 'Oceania',
    'Vanuatu': 'Oceania'
}

def str_para_bool(valor):
    return 1 if valor.upper() == 'TRUE' else 0

def extrair_ano_torneio(tournament):
    # Padrões comuns de anos em nomes de torneios
    padroes = [
        r'\b(19|20)\d{2}\b',  # Anos completos (1900-2099)
        r'\b\d{2}\b',         # Anos abreviados (assume 19xx para < 30, 20xx para >= 30)
    ]
    
    for padrao in padroes:
        match = re.search(padrao, tournament)
        if match:
            ano = match.group(0)
            # Se for ano abreviado, converte para completo
            if len(ano) == 2:
                ano_num = int(ano)
                if ano_num >= 30:
                    ano = f"19{ano}"
                else:
                    ano = f"20{ano}"
            return f"{ano}-01-01"
    
    # Mapeamento de eventos específicos para anos
    eventos_especificos = {
        'World Cup': '2022-01-01',  # Ajuste conforme necessário
        'Copa América': '2021-01-01',  # Ajuste conforme necessário
        'European Championship': '2020-01-01',  # Ajuste conforme necessário
        'Olympics': '2020-01-01'  # Ajuste conforme necessário
    }
    
    for evento, ano in eventos_especificos.items():
        if evento.lower() in tournament.lower():
            return ano
            
    return None

def inserir_torneio(tournament):
    try:
        ano = extrair_ano_torneio(tournament)
        sql_torneio = """
            INSERT IGNORE INTO Torneio (Nome, Ano) 
            VALUES (%s, %s)
            ON DUPLICATE KEY UPDATE Ano = COALESCE(Ano, VALUES(Ano))
        """
        cursor.execute(sql_torneio, (tournament, ano))
        db.commit()
    except mysql.connector.Error as err:
        print(f"Erro ao inserir torneio: {err}")

def determinar_continente(pais):
    return PAISES_CONTINENTES.get(pais, 'Desconhecido')

def inserir_selecao(home_team, away_team):
    try:
        sql_selecao = """
            INSERT IGNORE INTO Selecao (Nome, Continente) 
            VALUES (%s, %s), (%s, %s)
            ON DUPLICATE KEY UPDATE Continente = VALUES(Continente)
        """
        cursor.execute(sql_selecao, (
            home_team, determinar_continente(home_team),
            away_team, determinar_continente(away_team)
        ))
        db.commit()
    except mysql.connector.Error as err:
        print(f"Erro ao inserir seleção: {err}")

# [O resto do código permanece o mesmo...]

def carregar_csv(nome_arquivo, func_inserir, pre_func_inserir=None, jogador_inserir=None):
    try:
        with open(nome_arquivo, mode='r', encoding='utf-8') as arquivo:
            leitor_csv = csv.reader(arquivo)
            next(leitor_csv)  # Pular cabeçalho
            for linha in leitor_csv:
                if pre_func_inserir:
                    pre_func_inserir(*linha)
                if jogador_inserir and len(linha) >= 5:
                    jogador_inserir(linha[4], linha[3])
                func_inserir(*linha)
    except Exception as e:
        print(f"Erro ao carregar {nome_arquivo}: {e}")

def inserir_goleador(scorer, team):
    try:
        # Verifique se o jogador já existe na tabela JogadorMarcador
        sql_check_jogador = "SELECT COUNT(*) FROM JogadorMarcador WHERE Nome = %s"
        cursor.execute(sql_check_jogador, (scorer,))
        if cursor.fetchone()[0] == 0:
            # Se não existir, insira o novo jogador
            sql_insert_jogador = "INSERT INTO JogadorMarcador (Nome, Nacionalidade) VALUES (%s, %s)"
            # Determine a nacionalidade com base no nome da equipe ou outra lógica
            nacionalidade = determinar_nacionalidade(team)
            cursor.execute(sql_insert_jogador, (scorer, nacionalidade))
        
        # Insira a relação entre o jogador e a seleção
        sql_insert_selecao_jogador = "INSERT INTO SelecaoJogador (fk_Selecao_Nome, fk_JogadorMarcador_Nome) VALUES (%s, %s)"
        cursor.execute(sql_insert_selecao_jogador, (team, scorer))
        
        db.commit()
    except mysql.connector.Error as err:
        print(f"Erro ao inserir goleador: {err}")
        db.rollback()

def inserir_jogador(nome, equipe):
    try:
        # Verifique se o jogador já existe na tabela JogadorMarcador
        sql_check_jogador = "SELECT COUNT(*) FROM JogadorMarcador WHERE Nome = %s"
        cursor.execute(sql_check_jogador, (nome,))
        if cursor.fetchone()[0] == 0:
            # Se não existir, insira o novo jogador
            sql_insert_jogador = "INSERT INTO JogadorMarcador (Nome, Nacionalidade) VALUES (%s, %s)"
            # Determine a nacionalidade com base no nome da equipe ou outra lógica
            nacionalidade = determinar_nacionalidade(equipe)
            cursor.execute(sql_insert_jogador, (nome, nacionalidade))
        
        # Insira a relação entre o jogador e a seleção
        sql_insert_selecao_jogador = "INSERT INTO SelecaoJogador (fk_Selecao_Nome, fk_JogadorMarcador_Nome) VALUES (%s, %s)"
        cursor.execute(sql_insert_selecao_jogador, (equipe, nome))
        
        db.commit()
    except mysql.connector.Error as err:
        print(f"Erro ao inserir jogador: {err}")
        db.rollback()      

def inserir_partida(date, home_team, away_team, home_score, away_score, tournament, city, country, neutral):
    try:
        # Verifica se a partida já existe (mesmo com as seleções invertidas)
        check_sql = """
            SELECT COUNT(*) 
            FROM Partida 
            WHERE Data = %s 
            AND ((fk_Selecao_Nome1 = %s AND fk_Selecao_Nome2 = %s) 
                 OR (fk_Selecao_Nome1 = %s AND fk_Selecao_Nome2 = %s))
        """
        cursor.execute(check_sql, (date, home_team, away_team, away_team, home_team))
        if cursor.fetchone()[0] > 0:
            # Se existir, faz update
            update_sql = """
                UPDATE Partida 
                SET HomeScore = %s,
                    AwayScore = %s,
                    fk_Torneio_Nome = %s,
                    Cidade = %s,
                    Pais = %s,
                    Campo_neutro = %s
                WHERE Data = %s 
                AND ((fk_Selecao_Nome1 = %s AND fk_Selecao_Nome2 = %s) 
                     OR (fk_Selecao_Nome1 = %s AND fk_Selecao_Nome2 = %s))
            """
            valores_update = (
                home_score, away_score, tournament, city, country, 
                str_para_bool(neutral), date, home_team, away_team, away_team, home_team
            )
            cursor.execute(update_sql, valores_update)
        else:
            # Se não existir, insere novo
            insert_sql = """
                INSERT INTO Partida 
                (Data, fk_Selecao_Nome1, fk_Selecao_Nome2, HomeScore, 
                AwayScore, fk_Torneio_Nome, Cidade, Pais, Campo_neutro) 
                VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
            """
            valores_insert = (
                date, home_team, away_team, home_score, 
                away_score, tournament, city, country, str_para_bool(neutral)
            )
            cursor.execute(insert_sql, valores_insert)
        
        db.commit()
        print(f"Partida processada com sucesso: {date} - {home_team} vs {away_team}")
    
    except mysql.connector.Error as err:
        print(f"Erro ao processar partida: {err}")
        print(f"Dados: {date}, {home_team}, {away_team}")
        db.rollback()

# Carregar e inserir dados
carregar_csv(
    'results.csv',
    inserir_partida,
    pre_func_inserir=lambda date, home_team, away_team, home_score, away_score, tournament, city, country, neutral: (
        inserir_torneio(tournament),
        inserir_selecao(home_team, away_team)
    )
)

carregar_csv(
    'goalscorers.csv',
    inserir_goleador,
    jogador_inserir=lambda scorer, team: inserir_jogador(scorer, team)
)

# Atualizar anos faltantes na tabela Torneio
sql_update_anos = """
UPDATE Torneio t
JOIN (
    SELECT DISTINCT p.fk_Torneio_Nome, YEAR(p.Data) as ano
    FROM Partida p
    WHERE p.fk_Torneio_Nome IS NOT NULL
) p ON t.Nome = p.fk_Torneio_Nome
SET t.Ano = CONCAT(p.ano, '-01-01')
WHERE t.Ano IS NULL
"""

try:
    cursor.execute(sql_update_anos)
    db.commit()
except mysql.connector.Error as err:
    print(f"Erro ao atualizar anos dos torneios: {err}")

# Fechar conexões
cursor.close()
db.close()
```
